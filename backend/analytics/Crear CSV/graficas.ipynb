```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# GrÃ¡fico 1: Histograma de precios por categorÃ­a
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Precio', hue='Categoria', kde=True, bins=10)
plt.title('Histograma de precios por categorÃ­a')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# GrÃ¡fico 2: DistribuciÃ³n de edades de los clientes
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Edad', kde=True, bins=10)
plt.title('DistribuciÃ³n de edades de los clientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# GrÃ¡fico 3: GrÃ¡fico de conteo de gÃ©nero de los clientes
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sexo')
plt.title('GrÃ¡fico de conteo de gÃ©nero de los clientes')
plt.xlabel('GÃ©nero')
plt.ylabel('Cantidad')
plt.show()
```

```python
# GrÃ¡fico 4: GrÃ¡fico de conteo de compras por marca
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Marca')
plt.title('GrÃ¡fico de conteo de compras por marca')
plt.xlabel('Cantidad')
plt.ylabel('Marca')
plt.show()
```

```python
# Calcular los 20 perfumes mÃ¡s populares
top_20_perfumes = df['Nombre_Perfume'].value_counts().head(20).index

# Filtrar el DataFrame original para obtener solo los 20 mejores perfumes
df_top_20 = df[df['Nombre_Perfume'].isin(top_20_perfumes)]

# Crear el grÃ¡fico de barras para los 20 mejores perfumes
plt.figure(figsize=(10, 8))
sns.countplot(data=df_top_20, y='Nombre_Perfume', order=df_top_20['Nombre_Perfume'].value_counts().index)
plt.title('GrÃ¡fico de barras de los 20 mejores perfumes')
plt.xlabel('Cantidad')
plt.ylabel('Nombre del Perfume')
plt.show()
```

```python
# Convertir la columna de fecha a tipo datetime (si no estÃ¡ en este formato)
df['Fecha_Compra'] = pd.to_datetime(df['Fecha_Compra'])

# Crear el grÃ¡fico del historial de compra con la fecha
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Fecha_Compra', y='Precio', marker='o')
plt.title('Historial de compra con la fecha')
plt.xlabel('Fecha de compra')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filtrar el DataFrame por sexo hombre
df_hombre = df[df['Sexo'] == 'Hombre']

# Filtrar el DataFrame por sexo mujer
df_mujer = df[df['Sexo'] == 'Mujer']

# Crear el grÃ¡fico del historial de compra por fecha y sexo
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_hombre, x='Fecha_Compra', y='Precio', label='Hombre', marker='o')
sns.lineplot(data=df_mujer, x='Fecha_Compra', y='Precio', label='Mujer', marker='o')
plt.title('Historial de compra por fecha y sexo')
plt.xlabel('Fecha de compra')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend(title='Sexo')
plt.show()
```
