```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('perfumes.csv')
display(df)
```

```python
# Gráfico 1: Histograma de precios por categoría
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Precio', hue='Categoria', kde=True, bins=10)
plt.title('Histograma de precios por categoría')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico 2: Distribución de edades de los clientes
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Edad', kde=True, bins=10)
plt.title('Distribución de edades de los clientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico 3: Gráfico de conteo de género de los clientes
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sexo')
plt.title('Gráfico de conteo de género de los clientes')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico 4: Gráfico de conteo de compras por marca
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Marca')
plt.title('Gráfico de conteo de compras por marca')
plt.xlabel('Cantidad')
plt.ylabel('Marca')
plt.show()
```

```python
# Calcular los 20 perfumes más populares
top_20_perfumes = df['Nombre_Perfume'].value_counts().head(20).index

# Filtrar el DataFrame original para obtener solo los 20 mejores perfumes
df_top_20 = df[df['Nombre_Perfume'].isin(top_20_perfumes)]

# Crear el gráfico de barras para los 20 mejores perfumes
plt.figure(figsize=(10, 8))
sns.countplot(data=df_top_20, y='Nombre_Perfume', order=df_top_20['Nombre_Perfume'].value_counts().index)
plt.title('Gráfico de barras de los 20 mejores perfumes')
plt.xlabel('Cantidad')
plt.ylabel('Nombre del Perfume')
plt.show()
```

```python
# Convertir la columna de fecha a tipo datetime (si no está en este formato)
df['Fecha_Compra'] = pd.to_datetime(df['Fecha_Compra'])

# Crear el gráfico del historial de compra con la fecha
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Fecha_Compra', y='Precio', marker='o')
plt.title('Historial de compra con la fecha')
plt.xlabel('Fecha de compra')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filtrar el DataFrame por sexo hombre
df_hombre = df[df['Sexo'] == 'Hombre']

# Filtrar el DataFrame por sexo mujer
df_mujer = df[df['Sexo'] == 'Mujer']

# Crear el gráfico del historial de compra por fecha y sexo
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_hombre, x='Fecha_Compra', y='Precio', label='Hombre', marker='o')
sns.lineplot(data=df_mujer, x='Fecha_Compra', y='Precio', label='Mujer', marker='o')
plt.title('Historial de compra por fecha y sexo')
plt.xlabel('Fecha de compra')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend(title='Sexo')
plt.show()
```
